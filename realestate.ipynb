```
x_data = [2104, 1600, 2400]
y_data = [399.9, 329.9, 369]

x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
w1 = 0.2 # a random guess: random value
w2 = -34
b = 0.0
# our model forward pass
def forward(x):      #y prediction
  return ((pow(x,2) * w2) + (x * w1) + b)
             
             
# Loss function
def loss(x, y):
  y_pred = forward(x)
  return (y_pred - y) * (y_pred - y)

# compute gradient
def gradient1(x, y): # d_loss/d_w
  y_pred = forward(x)
  return ((2 * x) * (((pow(x,2) * w2) + (x * w1) + b) - y))

def gradient2(x, y): # d_loss/d_w
  y_pred = forward(x)
  return ((2 * pow(x,2)) * (((pow(x,2) * w2) + (x * w1) + b) - y))

def gradientb(x,y):
  return (2 * (((pow(x,2) * w2) + (x * w1) + b) - y))


# Before training
print("predict (before training)", 4, forward(4))


for epoch in range(100):
  grad1=0
  grad2=0
  gradb=0
  l=0
  for x_val, y_val in zip(x_data, y_data): 
    grad1 = gradient1(x_val, y_val)
    grad2 = gradient2(x_val, y_val)
    gradb = gradientb(x_val, y_val)
    l = l + loss(x_val, y_val)
    print("\tgrad: ", x_val, y_val, grad1, grad2, gradb)  
  w1 = w1 - (0.00000000000001 * grad1) 
  w2 = w2 - (0.00000000000001 * grad2)
  b = b - (0.00000000000001 * gradb)
  print("progress:", epoch, "w1=", w1, "w2=",w2, "loss=", l, "bias=",b)

```
